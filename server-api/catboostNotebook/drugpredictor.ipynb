```python
import pandas as pd
import numpy as np


df = pd.read_csv('pharmacy_dataset.csv')
df.head(10)
```

```python
import pandas as pd

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype in ['float64', 'int64']:
            mean_val = df[col].mean()
            df[col] = df[col].fillna(mean_val)
        elif df[col].dtype == 'object':
            df[col] = df[col].fillna('Missing')


# Display the first few rows of the preprocessed dataframe
df.head()
```

```python
import pandas as pd
df['monthID'] = pd.to_datetime(df['Месяц']).dt.month - 1
df.head()
```

```python
import pandas as pd

df['Месяц'] = pd.to_datetime(df['Месяц'])

min_date = df['Месяц'].min()
df['Месяц'] = (df['Месяц'].dt.year - min_date.year) * 12 + (df['Месяц'].dt.month - min_date.month)


df['ЖНВЛП'] = df['ЖНВЛП'].apply(lambda x: 1 if x == 'Да' else 0)
df['ПККН'] = df['ПККН'].apply(lambda x: 1 if x == 'Да' else 0)

df.head()
```

```python

from sklearn.model_selection import train_test_split, GridSearchCV
from catboost import CatBoostRegressor


features = df.drop('Кол-во', axis=1)
target = df['Кол-во']


X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


categorical_features_indices = [i for i, col in enumerate(X_train.columns) if X_train[col].dtype == 'object']


model = CatBoostRegressor(verbose=2, random_state=42)


param_grid = {
    'iterations': [400],
    'learning_rate': [0.1],
    'depth': [9],
}


grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train, cat_features=categorical_features_indices)

# Print best parameters and best score
print("Best parameters found: ", grid_search.best_params_)
print("Best negative mean squared error found: ", grid_search.best_score_)

# Get the best model
best_model = grid_search.best_estimator_
```

```python
pred = best_model.predict(X_test)
```

```python
# prompt: display 20 predict actual values in range

# Display the first 20 predicted vs actual values
print("Predicted vs Actual values for the first 20 samples:")
for i in range(20):
    display(X_test.iloc[i])
    print(f"Sample {i+1}: Predicted = {pred[i]:.2f}, Actual = {y_test.iloc[i]:.2f}")
```
